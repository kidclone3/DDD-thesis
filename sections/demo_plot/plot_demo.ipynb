```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_with_arrows(fig, ax, df, color="green", df2=None, df3=None, df4=None):
    x = df['x']
    y = df['y']

    def draw_plot(fig, ax, x, y, color):
        # calculate position and direction vectors:
        x0 = x.iloc[range(len(x) - 1)].values
        x1 = x.iloc[range(1, len(x))].values
        y0 = y.iloc[range(len(y) - 1)].values
        y1 = y.iloc[range(1, len(y))].values
        xpos = (x0 + x1) / 2
        ypos = (y0 + y1) / 2
        xdir = x1 - x0
        ydir = y1 - y0

        # plot the line, color is green, node is green,
        ax.plot(x, y, color=color, linewidth=0.5)

        # plot arrow on each line:
        for X, Y, dX, dY in zip(xpos, ypos, xdir, ydir):
            ax.annotate("", xytext=(X, Y), xy=(X + 0.001 * dX, Y + 0.001 * dY),
                        arrowprops=dict(arrowstyle="->", color=color), size=10)

        # Small points
        ax.scatter(x, y, color=color, s=20)

    draw_plot(fig, ax, x, y, color)
    if df2 is not None:
        x = df2['x']
        y = df2['y']
        draw_plot(fig, ax, x, y, color)

    if df3 is not None:
        x = df3['x']
        y = df3['y']
        draw_plot(fig, ax, x, y, color)
    if df4 is not None:
        x = df4['x']
        y = df4['y']
        draw_plot(fig, ax, x, y, color)


def gen_data_and_plot(list_df: list, list_color: list = None, selected_points: list = None):
    fig, ax = plt.subplots()

    # show grid
    ax.grid(True)

    ax.set_xlim(0, 5.1)

    # y axis from 1 to 4 with 1 step
    ax.set_yticks(range(1, 5, 1))

    # remove top border
    ax.spines['top'].set_visible(False)

    all_x = []
    all_y = []

    for i, data in enumerate(list_df):
        df = data
        x = df['x'].tolist()
        y = df['y'].tolist()
        all_x.extend(x)
        all_y.extend(y)
        df2 = pd.DataFrame({'x': [x[0], x[2]],
                            'y': [y[0], y[2]]})
        df3 = pd.DataFrame({'x': [x[1], x[3]],
                            'y': [y[1], y[3]]
                            })
        df4 = pd.DataFrame({'x': [x[0], x[3]],
                            'y': [y[0], y[3]]
                            })
        plot_with_arrows(fig, ax, df, list_color[i], df2, df3, df4)

    # Vẽ các vòng tròn màu tím cho tất cả các điểm
    ax.scatter(all_x, all_y, facecolors='none', edgecolors='purple', s=100)

    # Vẽ vòng tròn có dấu X cho các điểm được chọn
    if selected_points:
        selected_x = [all_x[i] for i in selected_points]
        selected_y = [all_y[i] for i in selected_points]
        ax.scatter(selected_x, selected_y, facecolors='none', edgecolors='purple', s=100)
        for (x, y) in zip(selected_x, selected_y):
            ax.plot([x - 0.1, x + 0.1], [y - 0.1, y + 0.1], color='purple')
            ax.plot([x - 0.1, x + 0.1], [y + 0.1, y - 0.1], color='purple')

    plt.xlabel('Time')
    plt.ylabel('Nodes')
    plt.show()


# ABSPT:(0,0)
# Data1
x = [0, 1.34, 1.76127, 2.5674]
y = [1, 2, 3, 4]
df1 = pd.DataFrame({
    'x': x,
    'y': y
})

# ABSPT:(0,1)
# Data3 (0,1), (1,1.66), (2,2.23607), (3,3.0826)
x = [1, 1.66, 2.23607, 3.0826]
y = [1, 2, 3, 4]
df3 = pd.DataFrame({
    'x': x,
    'y': y
})
# ABSPT:(1,2)
# Data5 (0,1.70833), (1,2), (2,2.74766), (3,3.63)
x = [1.70833, 2, 2.74766, 3.63]
y = [1, 2, 3, 4]
df5 = pd.DataFrame({
    'x': x,
    'y': y
})

# Data4 (0,2), (1,2.14), (2,3.00151), (3,3.9016)
x = [2, 2.14, 3.00151, 3.9016]
y = [1, 2, 3, 4]
df4 = pd.DataFrame({'x': x, 'y': y})

# Data2
x = [2.86862, 2.8957, 4.03846, 5]
y = [1, 2, 3, 4]
df2 = pd.DataFrame({
    'x': x,
    'y': y
})

# List of selected points (indices of points in the combined all_x and all_y lists)
selected_points = [0, 4, 8, 12, 16]

gen_data_and_plot([df1, df3, df5, df4, df2], ['green', 'black', 'black', 'black', 'blue'], selected_points)
```
